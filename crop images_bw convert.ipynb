```python
from PIL import Image
import numpy as np
import cv2
import os, sys
```

```python
def crop_img(START_PATH,DES_PATH):
    dirs = os.listdir(DES_PATH) #list of all the files in destination directory
    n = len(dirs) #number of all the image in destination directory
    all_img = os.listdir(START_PATH)
    for img_path in all_img:
        #print(img_path)
        img = Image.open(START_PATH+img_path)
        (width,height)=img.size
        size = 256 #size of extracted images
        col = width//size #number of col
        row = height//size #number of row
        for r in range(row):
            for c in range(col):
                n = n+1
                box = (size*c,size*r,size*c+size,size*r+size) #upper left and lower right of the image
                img_crop = img.crop(box)
                img_crop.save(DES_PATH+str(n)+'.png')
    #return n
```

```python
def blackandwhite(START_PATH,DES_PATH):
    dirs = os.listdir(START_PATH)
    target_dirs = os.listdir(DES_PATH)
    n = len(target_dirs)
    for img_path in dirs:
        n = n+1
        originalImage = cv2.imread(START_PATH+img_path)
        grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
        (thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY_INV)
        #cv2.imwrite(DES_PATH+str(n)+'.png', blackAndWhiteImage)
        cv2.imwrite(DES_PATH+img_path, blackAndWhiteImage)
```
